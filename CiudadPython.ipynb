```python
%pip install openrouteservice folium

from openrouteservice import convert
import openrouteservice
import folium
import json

client = openrouteservice.Client(key='5b3ce3597851110001cf62480f0b073e0b4f4a509451c6c945555492')

Punto_A = {
    "nombre": "UTP",
    "latitud": 9.022100303741224,
    "longitud": -79.53200963246745,
}

Punto_B = {
    "nombre": "Terminal de Albrook",
    "latitud": 8.971513053167264,
    "longitud": -79.55273002290262,
}

coordenadas = (
    [Punto_A["longitud"], Punto_A["latitud"]],
    [Punto_B["longitud"], Punto_B["latitud"]]
)

def estilo(mensaje):
  return f"<h1>{mensaje}</h1>"

res = client.directions(coordenadas)
geometry = client.directions(coordenadas)['routes'][0]['geometry']
decoded = convert.decode_polyline(geometry)

mensaje = f"Distancia: {str(round(res['routes'][0]['summary']['distance']/1000,1))} Kilometros - Duracion: {str(round(res['routes'][0]['summary']['duration']/60,1))} Minutos"

m = folium.Map(location=[Punto_A["latitud"], Punto_A["longitud"]],zoom_start=13, control_scale=True,tiles="cartodbpositron")
folium.GeoJson(decoded).add_child(folium.Popup(estilo(mensaje),max_width=300)).add_to(m)

folium.Marker(
    location=list(coordenadas[0][::-1]),
    popup=estilo(Punto_A["nombre"]),
    icon=folium.Icon(color="green"),
).add_to(m)

folium.Marker(
    location=list(coordenadas[1][::-1]),
    popup=estilo(Punto_B["nombre"]),
    icon=folium.Icon(color="red"),
).add_to(m)

m
```
