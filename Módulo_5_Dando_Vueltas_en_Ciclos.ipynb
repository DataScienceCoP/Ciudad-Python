## ¿Repetir?

Sí, con Python puedes repetir cosa varias veces y ejecutarlas, según tus condiciones, y hasta cuando tu quieras.

## Loops o "Iteraciones"

Las computadoras pueden repetir muchas acciones o cálculos muy rápidamente. 
Un loop es una secuencia de instrucciones de código que se ejecuta repetidas veces, hasta que la condición asignada a dicho bucle deja de cumplirse. 

### FOR
Este tipo de bucle repite un bloque de código un número específico de veces. Usualmente usamos el loop **for** con listas y cuando sabemos cuántas veces necesitamos repetir un bloque de código.


````python
HorarioBuses = [5, 6, 7, 8, 9, 10, 11, 12, 13]
for horas in HorarioBuses:
  print(horas)
````

#### Iterar con Bucles FOR

¿Qué hacemos si no queremos revisar todos los elementos? ¿Qué pasa si solo queremos iterar, digamos 3 veces? ¿O solo pasar por un rango específico de números? 

**for** -iterator- **in** -group of items-:

El -iterator- y el -group of items- son las partes que se reemplazas cuando se crea su propio bucle **for**. Entonces, si quisiera iterar 3 veces, en lugar de iterar a través de una lista completa o tupla con elementos en ella, reemplazaríamos el -iterator- y el -group of items- con lo siguiente:


````python
for i in range(3):
  print(i)
````

También, puede tomar otro parámetro que sea un punto de partida. Esto nos permite hacer algo similar a dividir listas y tuplas iterando sobre un rango específico de números. Entonces, si quisiéramos saltar directamente al número 10 y luego iterar a través de los números del 10 al 20, usaríamos la función range () de esta manera, usando dos parámetros:

````python
for i in range(10,21):
  print(i)
````

Entonces, si solo quisiéramos imprimir los múltiplos de 10 en los números entre 1 y 100, usaríamos la función range ()

````python
for i in range (0,110,10):
  print(i)
````

#### Bucle FOR Anidado
Lo bueno de los bluces de Python es que se pueden anidar, es decir, podemos usar uno o mas bucles dentro de otro bucle. Esto nos permite resolver problemas aun más complejos. Vamos a utilizar el bucle **for** anidado para imprimir el siguiente patrón:

````python
1
22
333
4444
55555
````

El ciclo externo itera a través del rango de 1 a 5 y para cada elemento en esa secuencia.
Entra en el ciclo interno donde itera sobre un rango de ese elemento.
Para cada iteración de ese artículo, imprime el artículo.
Solo abandona el bucle interno cuando ha iterado completamente a través de un rango de ese elemento.
Cuando sale del bucle interno, vuelve al bucle externo y el proceso continúa hasta que ha iterado completamente sobre su secuencia.

````python
for i in range(1,6):
    for j in range(0,i):
        print(i, end="")
    print("")
````

### WHILE

Este tipo de bucle también repite un bloque de código una y otra vez, pero seguirá repitiéndose mientras una expresión booleana siga siendo verdadera.
Solemos usar **while** cuando no sabemos cuántas veces necesitamos repetir un bloque de código. Recuerde que en el loop **for**, sabemos exactamente la cantidad de veces que debe repetirse un bloque de código.


````python
RutaAutobus = int(input("Seleccione la ruta que desea tomar: 1) Estandar, 2) Expreso o 3) Corredor= "))
while (RutaAutobus != 1 and RutaAutobus != 2 and RutaAutobus != 3):
  print("Opcion no valida. Intenta de nuevo!")
  RutaAutobus = int(input("Seleccione la ruta que desea tomar: 1) Estandar, 2) Expreso or 3) Corredor= "))

print("Entendido! Has seleccionada la opcion ", RutaAutobus)
````
