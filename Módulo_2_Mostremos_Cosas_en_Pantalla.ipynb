## Aprendamos un poco sobre los tipos de datos

En cualquier lenguaje de programación de alto nivel se manejan tipos de datos. Los tipos de datos definen un conjunto de valores que tienen una serie de características y propiedades determinadas.

Para saber el tipo de un dato en Python utilizamos la instrucción **type()**.

### Tipo Numero

En Python representamos los números según su tipo:

- Entero
- Flotante
- Complejo (Imaginario)
- Booleano

#### Enteros

Para un entero podemos escribir el número solamente y Python lo va a aceptar como válido:

```python
4
```

```python
type(1)
```

También los números pueden ser negativos, utilizando el símbolo '-'

```python
-1
```

#### Flotante

Para un flotante, nos estamos refiriendo a un número con punto decimal:

```python
2.7854
```

```python
type(3.1416)
```

#### Complejo (Imaginario)

Son aquellos números que se expresan como la suma de un número real y otro imaginario. Mientras que un número imaginario es un número que resulta de la raíz cuadrada de un número negativo.

Para un número imaginario (o complejo) colocamos la letra 'j' al lado del número:

```python
2j
```

```python
1+3j
```

```python
type(1+3j)
```

#### Tipo Booleano

Con el tipo booleano nos referimos al concepto de si algo es cierto o falso. En Python el tipo booleano principalmente se refiere a **True** o **False**

```python
True
```

```python
False
```

```python
type(True)
```

# Ejecución de Programas con Entradas

A continuación, vamos a crear nuestro primero programa. Preguntaremos el nombre al usuario, saludamos y decimos cuántas letras tiene su nombre. Más tarde, preguntamos su edad y calculamos cuántos años cumplirá el próximo año. Esta interacción se logra con la función **input()** la cual captura el nombre/edad proporcionado y las guardamos en una variable que llamamos **mi_nombre** y **mi_edad**, respectivamente. Veamos!

```python
# Mi primer programa dice hola y pregunta por tu nombre.

print('Hola Mundo!')
print('¿Cuál es tu nombre?')
mi_nombre = input()  
print('Es un placer conocerte, ' + mi_nombre)
print('Tu nombre tiene ',len(mi_nombre),' letras')

print('¿Cuál es tu edad?')
mi_edad = input()
print('Tendrás ' + str(int(mi_edad)+1) + ' el próximo año')

```
